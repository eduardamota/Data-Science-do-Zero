## Revisão MatplotLib 

```python
from matplotlib import pyplot as plt
```

```python
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2,543.3,1075.9,2862.5,5979.6,1028.7,14958.3]
```

```python
plt.plot(years, gdp, color = 'k', marker = 'o', linestyle = 'solid')
plt.title("Nominal GDP")
plt.show()
```

```python
movies = ["Annie Hall", "Duda", "Ale", "CasaBlanca", "Gandhi"]
num_oscars = [5, 11, 3, 8,10]

# plote as barras com coordenadas x à esquerda [0,1,2,3,4]
plt.bar(range(len(movies)), num_oscars)

plt.title("Filmes Favoritos")

plt.xticks(range(len(movies)),movies) #Rotulando com nome dos filmes
plt.show()
```
