```python
from collections import Counter 
```

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
```

```python
histogram = Counter(min(grade // 10 * 10,90)for grade in grades)
print(histogram)
```

No c√≥digo, a express√£o grade // 10 * 10 est√° arredondando cada nota para o m√∫ltiplo de 10 mais pr√≥ximo, garantindo que as notas sejam agrupadas em intervalos de 10. Isso √© feito usando a divis√£o inteira // para obter a parte inteira da divis√£o da nota por 10, e em seguida multiplicando por 10 para obter o m√∫ltiplo de 10 correspondente.

O valor resultante √© ent√£o comparado com 90 para garantir que o valor m√°ximo seja 90. Isso √© feito com a fun√ß√£o min() que retorna o valor m√≠nimo entre o valor calculado e 90. Isso √© √∫til para garantir que todas as notas superiores a 90 sejam agrupadas no mesmo intervalo.

Dessa forma, o c√≥digo cria um histograma que conta quantas notas est√£o em cada intervalo de 10, permitindo uma visualiza√ß√£o da distribui√ß√£o das notas em grupo

```python
plt.bar([x + 5 for x in histogram.keys()], #Movendo as barras para direita,
        histogram.values(),
        color='b',
        width=10, #Largura da barra
        edgecolor = (0,0,0)) #Escurecendo as bordas dos gr√°ficos 
         
plt.axis([-5,105,0,5])  #Definindo o limite dos eixos
plt.xticks([10 * i for i in range (11)])
plt.xlabel("Notas")
plt.ylabel ("N¬∫ de alunos")
plt.title("Distribui√ß√£o de Notas")
plt.show()
```

Usamos eixos mais sens√≠veis e conseguimos notar um resultado menos expressivo

```python
#Gr√°ficos de linha
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error = [x+y for x,y in zip(variance,bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, '-g', label='variance')
plt.plot(xs,bias_squared,'-r.',label = 'bias^2')
plt.plot(xs, total_error,'b:',label='total error')
plt.legend(loc=9)
plt.xlabel("Complexidade do modelo")
plt.xticks([])
plt.show()
```
