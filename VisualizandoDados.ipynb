```python
from collections import Counter 
```

```python
grades = [83,95,91,87,70,0,85,82,100,67,73,77,0]
```

```python
histogram = Counter(min(grade // 10 * 10,90)for grade in grades)
print(histogram)
```

No código, a expressão grade // 10 * 10 está arredondando cada nota para o múltiplo de 10 mais próximo, garantindo que as notas sejam agrupadas em intervalos de 10. Isso é feito usando a divisão inteira // para obter a parte inteira da divisão da nota por 10, e em seguida multiplicando por 10 para obter o múltiplo de 10 correspondente.

O valor resultante é então comparado com 90 para garantir que o valor máximo seja 90. Isso é feito com a função min() que retorna o valor mínimo entre o valor calculado e 90. Isso é útil para garantir que todas as notas superiores a 90 sejam agrupadas no mesmo intervalo.

Dessa forma, o código cria um histograma que conta quantas notas estão em cada intervalo de 10, permitindo uma visualização da distribuição das notas em grupo

```python
plt.bar([x + 5 for x in histogram.keys()], #Movendo as barras para direita,
        histogram.values(),
        color='b',
        width=10, #Largura da barra
        edgecolor = (0,0,0)) #Escurecendo as bordas dos gráficos 
         
plt.axis([-5,105,0,5])  #Definindo o limite dos eixos
plt.xticks([10 * i for i in range (11)])
plt.xlabel("Notas")
plt.ylabel ("Nº de alunos")
plt.title("Distribuição de Notas")
plt.show()
```

```python
mentions = [500,505]
years = [2017,2018]

plt.bar(years, mentions, 0.8)
plt.xticks(years)
plt.ticklabel_format(useOffset=False) #desativa o deslocamento automático
plt.axis([2016.5,2018.5,0,550])
```

Usamos eixos mais sensíveis e conseguimos notar um resultado menos expressivo

```python
#Gráficos de linha
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error = [x+y for x,y in zip(variance,bias_squared)]
xs = [i for i, _ in enumerate(variance)]

plt.plot(xs, variance, '-g', label='variance')
plt.plot(xs,bias_squared,'-r.',label = 'bias^2')
plt.plot(xs, total_error,'b:',label='total error')
plt.legend(loc=9)
plt.xlabel("Complexidade do modelo")
plt.xticks([])
plt.show()
```
